```python
from ann_visualizer.visualize import ann_viz
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
from keras.models import model_from_json
import tensorflow as tf
import re
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
np.random.seed(7)

dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")

X = dataset[:,0:8]
Y = dataset[:,8]

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X, Y, epochs=150, batch_size=10)

scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
class graph_neuron: #lembrando que é dirigido, mas comentar sobre backpropagation

  def __init__(self, keras_model) :

    if not isinstance(keras_model, Sequential):

            raise ValueError("Argumento keras_model deve ser do tipo keras.engine.sequential.Sequential.")

    self.keras_model = keras_model

    self.keras_model.summary()


  def generate_graph_visualization_PUBLICO(self, title: str):

    ann_viz(self.keras_model, title= title)


  def generate_adjacency_list_vbias_register_PRIVATE(self, dataframe_sem_bias):

    all_weights = []


    for layer in model.layers:

        if len(layer.get_weights()) > 0:

            weights = layer.get_weights()[1]

            all_weights.append(weights)

    all_weights_flat = np.concatenate(all_weights)

    ultimo_indice = dataframe_sem_bias.index[-1]

    valor_numerico_ultimo_indice = int(re.search(r'\d+', ultimo_indice).group())

    adjacency_list = []

    print(self.keras_model.layers[0].input_shape[1], valor_numerico_ultimo_indice)

    adjacency_nodes = [f'v{i}' for i in range(self.keras_model.layers[0].input_shape[1], valor_numerico_ultimo_indice + 1) if all_weights_flat[i - self.keras_model.layers[0].input_shape[1]] != 0]

    adjacency_list.append(adjacency_nodes)

    df = pd.DataFrame(adjacency_list, index=['vBias'])

    print(df)

    return df


  def generate_adjacency_list_each_layer_PRIVATE(self, index_of_layer, integer):

    if len(self.keras_model.layers) == index_of_layer:

      return pd.DataFrame([np.nan], index= [f"v{integer}"])

    adjacency_list = []

    for j in range(self.keras_model.layers[index_of_layer].input_shape[1]):

        adjacent_nodes = [f"v{i + self.keras_model.layers[index_of_layer].input_shape[1] + integer}" for i in range(self.keras_model.layers[index_of_layer].output_shape[1])]

        adjacency_list.append(adjacent_nodes)


    df = pd.DataFrame(adjacency_list, index= [f'v{k + integer}' for k in range(self.keras_model.layers[index_of_layer].input_shape[1])])

    return df


  def generate_adjacency_list_PUBLICO(self):

    track_of_vertice_count = 0

    dataframe_list = []

    for i in range(len(self.keras_model.layers) + 1):

      dataframe_temporary = self.generate_adjacency_list_each_layer_PRIVATE(i, track_of_vertice_count)

      dataframe_list.append(dataframe_temporary)

      if i == len(self.keras_model.layers):

        break

      track_of_vertice_count += self.keras_model.layers[i].input_shape[1]

    dataframe_sem_bias =  pd.concat(dataframe_list)

    dataframe_apenas_bias = self.generate_adjacency_list_vbias_register_PRIVATE(dataframe_sem_bias)

    df = pd.concat([dataframe_sem_bias, dataframe_apenas_bias]).fillna("")

    self.adjacency_list = df.astype(str)

    return self.adjacency_list


  def check_vertice_outdegree_4_list_PUBLIC(self, which_vertice):

    non_zero_values = self.adjacency_list.iloc[which_vertice, :] != ''

    if non_zero_values.sum() == 0:

      return 'Esse vértice possui grau de saída nulo. Se trata de um sumidouro.'

    return f'Esse vértice possui grau de saída igual a: {non_zero_values.sum()}'


  def check_4_adjacency_vertices_using_adjacency_list_PUBLIC(self, which_vertice1, which_vertice2):

    if which_vertice1 == f'v{len(self.adjacency_list) - 1}':

        which_vertice1 = f'vBias'

    if which_vertice2 == f'v{len(self.adjacency_list) - 1}':

        which_vertice2 = f'vBias'

    if f'v{which_vertice1}' in (self.adjacency_list.iloc[which_vertice2, :]).values or f'v{which_vertice2}' in (self.adjacency_list.iloc[which_vertice1, :]).values:

      return 'São adjacentes.'

    return 'Não são adjacentes'


  def generate_bias_vector_4_adjacency_matrix_PRIVATE(self, dataframe):

    biases_values = []

    for layer in self.keras_model.layers:
        biases = layer.get_weights()[1]
        biases_values.append(biases)

    flattened_biases = [item for sublist in biases_values for item in sublist]

    flattened_biases = [0] * 8 + flattened_biases

    new_dataframe = pd.DataFrame({'vBias': flattened_biases})

    new_dataframe = new_dataframe.T

    new_dataframe.columns = dataframe.columns

    new_dataframe = new_dataframe.applymap(lambda x: int(x) if x == 0 else x)

    dataframe = dataframe.append(new_dataframe, ignore_index=False)

    return dataframe


  def generate_adjacency_matrix_each_layer_PRIVATE(self, index_of_layer, integer):

    if len(self.keras_model.layers) == index_of_layer:

      return pd.DataFrame([np.zeros(integer, int)], index= [f"v{integer}"])

    adjacency_list = []

    for j in range(self.keras_model.layers[index_of_layer].input_shape[1]):


      adjacent_nodes = np.concatenate(
    (
        np.zeros(
            sum(
                layer.input_shape[1] if hasattr(layer, 'input_shape') and layer.input_shape is not None else 0
                for layer in self.keras_model.layers[:index_of_layer + 1]
            ),
            int
        ),
        [
            f"{value} * input"
            for count, value in enumerate (self.keras_model.layers[index_of_layer].get_weights()[0][j])
        ]
    ),
    axis=0
)

      adjacency_list.append(adjacent_nodes)



    df = pd.DataFrame(adjacency_list, index= [f'v{k + integer}' for k in range(self.keras_model.layers[index_of_layer].input_shape[1])])

    return df


  def generate_adjacency_matrix_PUBLICO(self):

    track_of_vertice_count = 0

    dataframe_list = []

    for i in range(len(self.keras_model.layers) + 1):

      dataframe_temporary = self.generate_adjacency_matrix_each_layer_PRIVATE(i, track_of_vertice_count)

      dataframe_list.append(dataframe_temporary)

      if i == len(self.keras_model.layers):

        break

      track_of_vertice_count += self.keras_model.layers[i].input_shape[1]

    df = pd.concat(dataframe_list)

    df.columns = ['v' + str(col) for col in df.columns]

    df = self.generate_bias_vector_4_adjacency_matrix_PRIVATE(df)

    df['vBias'] = ['0' for _ in range(len(df))]

    df = df.fillna(0).applymap(lambda x: round(x, 1) if x == 0 else x)

    self.adjacency_matrix = df.astype(str)

    return self.adjacency_matrix


  def check_vertice_outdegree_4_matrix_PUBLIC(self, which_vertice):

    non_zero_values = self.adjacency_matrix.iloc[which_vertice, :] != '0'

    if non_zero_values.sum() == 0:

      return 'Esse vértice possui grau de saída nulo. Se trata de um sumidouro.'

    return f'Esse vértice possui grau de saída igual a: {non_zero_values.sum()}'


  def check_vertice_indegree_4_matrix_PUBLIC(self, which_vertice):

    non_zero_values = self.adjacency_matrix.iloc[:, which_vertice] != '0'

    return f'Esse vértice possui graude de entrada igual a: {non_zero_values.sum()}'


  def check_4_adjacency_vertices_using_adjacency_matrix_PUBLIC(self, which_vertice1, which_vertice2):

    if self.adjacency_matrix.iloc[which_vertice2, which_vertice1] != '0' or self.adjacency_matrix.iloc[which_vertice1, which_vertice2] != '0':

      return 'São adjacentes.'

    return 'Não são adjacentes.'

  def check_anti_base_is_anti_root_PUBLIC(self):

    if self.keras_model.layers[-1].output_shape[1] == 1:

      return 'A anti-base desse digrafo é uma anti-raiz'

    return 'A anti-base desse digrafo não é uma raiz'
```

**SEGUINDO O DE CIMA:
Classificação Binária: Se você está resolvendo um problema de classificação binária, com duas classes mutuamente exclusivas (por exemplo, sim ou não, positivo ou negativo), geralmente utiliza-se um único neurônio na camada de saída com uma função de ativação sigmoid. O neurônio pode ser interpretado como a probabilidade de pertencer à classe positiva.

**SEGUINDO O DE CIMA: Classificação Multiclasse: Para problemas de classificação com mais de duas classes, você normalmente usa um número de neurônios na camada de saída igual ao número de classes. Cada neurônio representa uma classe e usa a função de ativação softmax para calcular probabilidades. Isso permite que a rede modele a probabilidade de pertencer a cada classe.

**SEGUINDO O DE CIMA: Regressão: Se você está realizando uma tarefa de regressão, onde está prevendo um valor numérico, geralmente usa um único neurônio na camada de saída sem aplicar uma função de ativação específica. O valor de saída é um número contínuo que representa a previsão.
